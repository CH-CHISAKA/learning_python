# Pandas: Getting Started

*Pandas* - This is a Python library used for working with datasets.

- It does by helping us make sense of *DataFrames*

*DataFrames* - A form of two-dimensional structured data, like a table with columns and rows. 

## Importing Data | CSV Files

*CSV* - Common Separated Values, and it's a file type that allows data to be saved in a table.

* Data presented in a tabular format is called *structured data*

```python
# Using Dictionaries to create a dataset of students and their grades

import pandas as pd

# Create a dictionary with student names as keys and their grades as values
student_grades = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 90,
    "Eve": 88
}   
# Convert the dictionary to a pandas DataFrame
pd.DataFrame.from_dict(student_grades, orient='index',columns=['Grade'])

# # Create a DataFrame from the dictionary
# df = pd.DataFrame(list(student_grades.items()), columns=['Student', 'Grade'])

# # Display the DataFrame
# print(df)
```

## Using JSON library 

```python
# Using JSON library to parse a JSON string into a Python dictionary

import json

info = """ {
    "firstName": "Alice",
    "lastName": "Smith",
    "age": 30,
    "city": "Stockholm",
    "hobbies": ["reading", "traveling", "cooking"]
}"""

# Parse JSON string into Python dictionary
info_dict = json.loads(info)

info_dict
```

```python
info_dict["city"]
```

## Loading data from Compressed Files

In big data ere - it is likely to read data from compressed files.

One way to unzip the data is to use gzip. 

```python
import gzip 
import json

with gzip.open('data/poland-bankruptcy-data-2008.json.gz', 'rt') as f:
    poland_data = json.load(f)
```

```python
# Check the type of poland_data
# type(poland_data)

# Check the keys in the poland_data dictionary
poland_data.keys()
```

```python
# Using the `from_dict`  Pandas function to read the data

# DataFrame()
df = pd.DataFrame.from_dict(poland_data['data'])

df.head()
```
