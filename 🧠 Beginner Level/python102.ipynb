# Strings

When handling datasets, it’s common to encounter files and folders named according to a consistent pattern. For instance, property price data for Colombia and Mexico might be organized like this:

```shell
1. colombia-real-estate-1.csv
2. colombia-real-estate-2.csv
3. colombia-real-estate-3.csv
4. mexico-city-real-estate-1.csv
5. mexico-city-real-estate-2.csv
6. mexico-city-real-estate-3.csv
7. mexico-city-real-estate-4.csv
8. mexico-city-real-estate-5.csv
9. mexico-city-test-features.csv
10. mexico-city-test-labels.csv
```

With a small number of files, it’s easy to manually locate the ones you need. However, as the number grows, it becomes more efficient to filter them programmatically. 

For example, if you only want files related to Mexico City, you can search for filenames that start with mexico-city-real-estate-.

This is where the .glob method comes in. It enables pattern-based file searching. For example, the pattern [0-9] matches any single digit from 0 through 9. 

There are several other commonly used patterns as well:

```shell
- * matches any sequence of characters
- ? matches exactly one character
- [a-z] matches any lowercase letter
- [A-Z] matches any uppercase letter
- [!a-z] matches any character that is not a lowercase letter
```
Using these patterns, you can efficiently retrieve all files associated with Mexico City by constructing an appropriate `.glob query`.



````python
import glob
````

````python
glob.glob("./data/*")
````

````python
display(glob.glob("./data/*"))
````

````python
glob.glob("./data/mexico*")
````

# For Loops

Used to iterate over a sequence (either a list, tuple, dictionary, set or string)
- One can execute a set of items, once for each item in a list, tuple, set etc. 

````python
# List of names
names = ["Jane Doe", "John Smith", "Alice Johnson", "Bob Brown", "Michael Jordan"]

for name in names:
    print(f"My name is {name}.")
````

````python
#  Looping letters in a name
fname = "Jane Doe"
for letter in fname:
    print(letter)
````

## Break Statement

Use the `break` statement to stop the loop before it has looped through all items

````python
for name in names:
    print(f"My name is {name}.")
    if name == "Bob Brown":
        break
    
# See the difference between the two loops above. The first loop will print all the names in the list, while the second loop will stop printing names once it reaches "Bob Brown".

# for name in names:
#     if name == "Bob Brown":
#         break
#     print(f"My name is {name}.")
````

## Return Values

Functions can send data back to the code that called them using the `return` statement

- When a function reaches a `return` statement, it stops executing and sends the results back

````python
# Function with return statement
def third_function():
    return "This is my third function in Python 102."

result = third_function()
print(result)
````

## Python Function Parameters
In programming, *parameters* are inputs you define for a function

* The variable listed inside the parentheses in the function definition.

````python
def greet(name): # Function with parameters -> name is the parameter
    print(f"Hello, {name}!")

greet("Alice") # Calling the function with an argument -> "Alice" is the argument
greet("Bob") # Calling the function with another argument -> "Bob" is the argument
````

## Python Function Arguments

Arguments - information passed in a function when it is called.

* Arguments are specified after the function name, inside the parentheses. -> Value passed to the function
* One can add as many arguments as they want, just separate them with a comma. 

````python
# Function with arguments 
def fruits(fruit_name): 
    print(f"Buy three, {fruit_name}!")

fruits("Bananas") # Calling the function with an argument -> "Banana" is the argument
fruits("Apples") # Calling the function with another argument -> "Apples" is the argument
````

###### Note

Function with multiple parameters and arguments
* By default, a function must be called with the correct number of arguments
* If the function expects 2 arguments, one must call it with exactly 2 arguments


````python

def name_and_age(name, age):
    print(f"My name is {name} and I am {age} years old.")

name_and_age("Jane Doe", 30)
name_and_age("John Smith", 25)
````

###### Note 
Default Parameter Value

One can assign default values to parameters. If the function is called without an argument, it uses the default value.

````python
def name_age_and_location(name, age, location = "Kenya"):
    print(f"My name is {name}, I am {age} years old and I live in {location}.")

name_age_and_location("Jane Doe", 30)
name_age_and_location("John Smith", 25, "Mexico")
````

## Working With Errors || Error handling 

Error handling ensure that code runs smoothly even with edge cases. 
* `try` and `expect` are the syntax used in error handling 

* The `try` block lets one test a block of code for errors.
* The `except` block lets one handle the error.

* `else` block lets one execute code when there is no error
* `finally` block lets one execute code, regardless of the result of the try-and except blocks.

````python
# # `try` will generate an exception because country is not defined
# try:
#     print(country)
# except NameError:
#     print("The variable 'country' is not defined. Please define it before using it.")
    
# # Alternatively, you can define the variable before using it to avoid the exception:
try:
    print(country)
except:
    print("The variable 'country' is not defined. Please define it before using it.")
````

````python
# One can define as many exception blocks as needed to handle different types of exceptions. For example:
try:
    print(country)
except NameError:
    print("The variable 'country' is not defined. Please define it before using it.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
````

### Else 

One can use the `else` keyword to define a block of code to be executed if no errors were raised.

````python
try:
    print("Hello!")
except:
    print("An error occurred while trying to print 'Hello!'")
else:
    print("Nothing went wrong, the code executed successfully.")
````

## Creating files using Context Manager 
This allows one to allocate and release resources precisely when one wants to.

````python
import os

os.makedirs("demo", exist_ok=True)

'''
os.makedirs("demo", exist_ok=True) : creates the folder if it doesn’t exist.
exist_ok=True prevents errors if it already exists.
'''

with open("demo/practice.txt", "w") as file:
    file.write("This is a practice file for Python 102.")
````
