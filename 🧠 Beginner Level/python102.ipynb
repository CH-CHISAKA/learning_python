# Strings

When handling datasets, it’s common to encounter files and folders named according to a consistent pattern. For instance, property price data for Colombia and Mexico might be organized like this:

```shell
1. colombia-real-estate-1.csv
2. colombia-real-estate-2.csv
3. colombia-real-estate-3.csv
4. mexico-city-real-estate-1.csv
5. mexico-city-real-estate-2.csv
6. mexico-city-real-estate-3.csv
7. mexico-city-real-estate-4.csv
8. mexico-city-real-estate-5.csv
9. mexico-city-test-features.csv
10. mexico-city-test-labels.csv
```

With a small number of files, it’s easy to manually locate the ones you need. However, as the number grows, it becomes more efficient to filter them programmatically. 

For example, if you only want files related to Mexico City, you can search for filenames that start with mexico-city-real-estate-.

This is where the .glob method comes in. It enables pattern-based file searching. For example, the pattern [0-9] matches any single digit from 0 through 9. 

There are several other commonly used patterns as well:

```shell
- * matches any sequence of characters
- ? matches exactly one character
- [a-z] matches any lowercase letter
- [A-Z] matches any uppercase letter
- [!a-z] matches any character that is not a lowercase letter
```
Using these patterns, you can efficiently retrieve all files associated with Mexico City by constructing an appropriate `.glob query`.



````python
import glob
````

````python
glob.glob("./data/*")
````

````python
display(glob.glob("./data/*"))
````

````python
glob.glob("./data/mexico*")
````

# Working with f-Strings

F-string allows for formatting of selected parts of a string

To specify a string as an f-string, by placing an `f` in front of the string literal

````python
name = f"My name is Jane Doe."
print(name)
````

````python
age = 30
f"My name is Jane Doe, {age} years old."
````

````python
location = "Mexico"
print(f"My name is Jane Doe and I live in {location}.")
````

# For Loops

Used to iterate over a sequence (either a list, tuple, dictionary, set or string)
- One can execute a set of items, once for each item in a list, tuple, set etc. 

````python
# List of names
names = ["Jane Doe", "John Smith", "Alice Johnson", "Bob Brown", "Michael Jordan"]

for name in names:
    print(f"My name is {name}.")
````

````python
#  Looping letters in a name
fname = "Jane Doe"
for letter in fname:
    print(letter)
````

## Break Statement

Use the `break` statement to stop the loop before it has looped through all items

````python
for name in names:
    print(f"My name is {name}.")
    if name == "Bob Brown":
        break
    
# See the difference between the two loops above. The first loop will print all the names in the list, while the second loop will stop printing names once it reaches "Bob Brown".

# for name in names:
#     if name == "Bob Brown":
#         break
#     print(f"My name is {name}.")
````
