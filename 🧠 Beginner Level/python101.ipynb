## Order of Operation
Defines the sequence in which different parts of a mathematical or logical expression are evaluated. Python generally follows the standard mathematical rules, often remembered by the acronym **PEMDAS**. 


The **PEMDAS** Hierarchy
For arithmetic, Python prioritizes operations in this specific order: 

* **Parentheses* () : Expressions inside parentheses are always evaluated first.

* **Exponents* ** : Powers (e.g.,) are calculated next.

* **MD (Multiplication & Division)* *, /, //, % : These share the same level of importance.

* **AS (Addition & Subtraction)* +, - : These are the last arithmetic operations performed. 


```python
# Example of order of operations
print((6+4*2)*2)
```

```python
# More complex order of operations
print((6+4*2)**2/2-3)
```
