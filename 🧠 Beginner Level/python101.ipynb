## Order of Operation
Defines the sequence in which different parts of a mathematical or logical expression are evaluated. Python generally follows the standard mathematical rules, often remembered by the acronym **PEMDAS**. 


The **PEMDAS** Hierarchy
For arithmetic, Python prioritizes operations in this specific order: 

* **Parentheses* () : Expressions inside parentheses are always evaluated first.

* **Exponents* ** : Powers (e.g.,) are calculated next.

* **MD (Multiplication & Division)* *, /, //, % : These share the same level of importance.

* **AS (Addition & Subtraction)* +, - : These are the last arithmetic operations performed. 


````python
# Example of order of operations
print((6+4*2)*2)
````

````python
# More complex order of operations
print((6+4*2)**2/2-3)
````

# Data Types
Data types are classes that classify items and determine what operations can be performed on them. 

- Because Python is **dynamically typed*, you don't need to declare a variable's type; it is automatically assigned based on the value you provide.

In Python, data is classified into distinct categories, each defining the operations that can be performed on it.

#### Mapping & Set Types

Used for storing data in ways that allow for fast retrieval or unique values. 

* dict: Stores data in key-value pairs, allowing values to be accessed by a unique key.

* set: An unordered collection of unique items (no duplicates).

* frozenset: An immutable version of a set.


#### Binary Types

Specialized for handling raw binary data like images or network packets. 

* bytes: An immutable sequence of bytes.

* bytearray: A mutable version of bytes.

* memoryview: Allows accessing an object's internal data without copying it.

### Type Conversion
This is all about changing one data type into another. 

It’s essential when you want to perform operations between incompatible types, or when you need input/output in a specific format.

Examples 
* String to integer

*  Integer to string

* String to float

````python
int_3 = 200
float26 = 6.5

# Converting int to float
float_from_int = float(int_3)
print(f"Converted int to float: {float_from_int} (type: {type(float_from_int)})")

# Converting float to int (note: this will truncate the decimal part)
int_from_float = int(float26)
print(f"Converted float to int: {int_from_float} (type: {type(int_from_float)})")

# Converting int to complex
complex_from_int = complex(int_3)
print(f"Converted int to complex: {complex_from_int} (type: {type(complex_from_int)})")

````

#### Aggregating Lists
Combining multiple lists into one or calculating a single summary value (like a sum or average) from a list's elements.

````python
# Using the + operator (concatenation)
# This creates a new list with all elements of the lists combined:
list_1 = [1, 2, 3]
list_2 = [4, 5, 6]
combined_list = list_1 + list_2
print(combined_list) # Output: [1, 2, 3, 4, 5, 6]
````

````python
# Using extend() (modifies the original list)
# This adds all elements from one list to another in-place:
list_1.extend(list_2)
print(list_1) # Output: [1, 2, 3, 4, 5, 6]
````

#### Zipping Items
This is a way to pair elements from two or more lists (or any iterables) together. 

- Think of it like a zipper that connects items side by side.

Example
```shell 
list1 = ['apple', 'banana', 'cherry']
list2 = [1, 2, 3]

zipped = zip(list1, list2)
print(list(zipped))  
# Output: [('apple', 1), ('banana', 2), ('cherry', 3)]
```

* Each element at the same index from list1 and list2 is combined into a tuple.

Note

zip() stops at the shortest list if the lists are not the same length.

You can zip any number of lists together.

The result is always tuples.

````python
# Creating a zip from lists
fruits = ["apple", "banana", "cherry"]
quantity = [10, 20, 30]
price_in_sek = [25, 15, 30]
# Using zip to combine the lists into a list of tuples
# zipped_list = list(zip(fruits, quantity, price_in_sek))
zipped_list = list(zip(fruits, quantity, price_in_sek))
print("The list consists of fruits, quantity, and price:", zipped_list)
````

#### Unzipping Items

You can also reverse a zip using the * operator:

````python
# Unzipping the list of tuples back into separate lists using the zip() function with the unpacking operator (*):
my_fruits, my_quantity, my_price = zip(*zipped_list)
print("Fruits:", my_fruits)
print("Quantity:", my_quantity)
print("Price in SEK:", my_price)
````
